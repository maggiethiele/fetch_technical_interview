```python
# SECOND: Provide SQL Queries Close-ended Questions

# ASSUMPTIONS: 
### products has duplicates, so we need to remove dupes by grabbing a distinct list before joining
### Every year has 365 days (to calculate year from days in USER_AGE)
### Every Month has 30 days (to calculate month from days in ACCOUNT_AGE)
### Receipts Scanned = COUNT(RECEIPT_ID)
### Sales = SUM(FINAL_SALE)

# 2a: What are the top 5 brands by receipts scanned among users 21 and over?
### ANSWER: There are 8 brands (and a NULL brand) in the ranking, indicating potential data quality issues in the products table (NULL BRAND). Specifically:
### tied for first at 6 units: NULL, DOVE and NERDS CANDY, which indicates that the products table needs some data cleanup to populate the Brand column more consistently.
### tied for fourth at 4 units: COCA-COLA, GREAT VALUE, HERSHEY'S, MEIJER, SOUR PATCH KIDS, and TRIDENT.
### *** If we remove NULL, DOVE and NERDS CANDY would tie for first and COCA-COLA, GREAT VALUE, HERSHEY'S, MEIJER, SOUR PATCH KIDS, and TRIDENT would tie for third.
two_a_query = """

----------------------------------------
-- Distinct list of BARCODES by BRAND to remove dupes
----------------------------------------
WITH distinct_barcode_by_brand AS (
    SELECT DISTINCT
       BARCODE
      ,BRAND
    FROM products
),

----------------------------------------
-- Create AGE by comparing Current Date with BIRTH_DATE
----------------------------------------
users_with_age AS (
    SELECT 
       u.ID
      -- Working out of SQL Lite, syntax for TSQL/MySQL is DATEDIFF(year, u.BIRTH_DATE, GETDATE()) 
      ,FLOOR((julianday(CURRENT_DATE) - julianday(u.BIRTH_DATE)) / 365) AS USER_AGE 
    FROM users u
),

----------------------------------------
-- Count receipts by Brand for users over 21 yrs old
----------------------------------------
receipts_by_brand AS (
    SELECT 
       p.BRAND
      ,COUNT(RECEIPT_ID) AS RECEIPTS
    FROM transactions t
      -- Only return transactions that have a mapping to products
      INNER JOIN distinct_barcode_by_brand p  ON t.BARCODE = p.BARCODE
      -- Inner Join to return only users that are 21 and over
      INNER JOIN users_with_age u             ON t.USER_ID = u.ID
                                             AND u.USER_AGE >= 21
    GROUP BY p.BRAND
)

SELECT 
   BRAND
  ,RECEIPTS
  ,RANK() OVER (ORDER BY RECEIPTS DESC) AS RANK
FROM receipts_by_brand
--WHERE BRAND IS NOT NULL

"""
# 2b: What are the top 5 brands by sales among users that have had their account for at least six months?
### ANSWER: 
### first: CVS at $72.00
### second: TRIDENT at $46.72
### third: DOVE at $42.88
### forth: COORS LIGHT at $34.96
### fifth: INCLUDING NULL - NULL at $16.65
###        EXCLUDING NULL - QUAKER at $16.60
two_b_query = """

----------------------------------------
-- Distinct list of BARCODES by BRAND to remove dupes
----------------------------------------
WITH distinct_barcode_by_brand AS (
    SELECT DISTINCT
       BARCODE
      ,BRAND
    FROM products
),

----------------------------------------
-- Generate ACCOUNT_AGE by comparing Current Date with CREATED_DATE
----------------------------------------
users_with_account_age AS (
    SELECT 
       u.ID
     ,CREATED_DATE
      -- Working out of SQL Lite, syntax for TSQL/MySQL is DATEDIFF(month, u.CREATED_DATE, GETDATE()) 
      -- Dividing by 30 since each month has roughly 30 days
     ,(julianday(CURRENT_DATE) - julianday(u.CREATED_DATE)) / 30 AS ACCOUNT_AGE 
    FROM users u
),

----------------------------------------
-- Sum sales by Brand for users with an account over 6 months old
----------------------------------------
sales_by_brand AS (
    SELECT 
       p.BRAND
      ,SUM(FINAL_SALE) AS SALES
    FROM transactions t
      -- Only return transactions that have a mapping to products
      INNER JOIN distinct_barcode_by_brand p  ON t.BARCODE = p.BARCODE
      -- Inner Join to return only users that are 21 and over
      INNER JOIN users_with_account_age u     ON t.USER_ID = u.ID
                                             AND u.ACCOUNT_AGE >= 6
    GROUP BY p.BRAND
    ORDER BY SUM(FINAL_SALE) DESC
)

SELECT 
   BRAND
  ,SALES
  ,RANK() OVER (ORDER BY SALES DESC) AS RANK
FROM sales_by_brand
--WHERE BRAND IS NOT NULL

"""
# 2c: What is the percentage of sales in the Health & Wellness category by generation?
### ANSWER:
### Baby Boomers = 49.94%
### Gen X = 21.88%
### Millennials = 31.18%
two_c_query = """

----------------------------------------
-- Distinct list of BARCODES to remove dupes filtered to just Health & Wellness
----------------------------------------
WITH distinct_barcode_health_wellness AS (
    SELECT DISTINCT
       p.BARCODE
      ,p.CATEGORY_1 
    FROM products p
    WHERE p.CATEGORY_1 = 'Health & Wellness' 
),

----------------------------------------
-- Create GENERATION based on birth year
----------------------------------------
users_with_generation AS (
    SELECT 
       u.ID
      ,CAST(strftime('%Y', u.BIRTH_DATE) AS INT) AS BIRTH_YEAR
      ,CASE WHEN CAST(strftime('%Y', u.BIRTH_DATE) AS INT) < 1946 THEN 'Silent Generation'
            WHEN CAST(strftime('%Y', u.BIRTH_DATE) AS INT) BETWEEN 1946 AND 1964 THEN 'Baby Boomers'
            WHEN CAST(strftime('%Y', u.BIRTH_DATE) AS INT) BETWEEN 1965 AND 1980 THEN 'Gen X'
            WHEN CAST(strftime('%Y', u.BIRTH_DATE) AS INT) BETWEEN 1981 AND 1996 THEN 'Millennials'
            WHEN CAST(strftime('%Y', u.BIRTH_DATE) AS INT) BETWEEN 1997 AND 2012 THEN 'Gen Z'
            WHEN CAST(strftime('%Y', u.BIRTH_DATE) AS INT) >= 2013 THEN 'Gen Alpha'
            ELSE NULL END AS GENERATION
    FROM users u
),

----------------------------------------
-- Sum sales by Generation for Health & Wellness category
----------------------------------------
sales_health_wellness AS (
    SELECT 
       u.GENERATION
      ,p.CATEGORY_1 
      ,SUM(FINAL_SALE) AS HEALTH_WELLNESS_SALES
    FROM transactions t
      INNER JOIN distinct_barcode_health_wellness p ON t.BARCODE = p.BARCODE
      INNER JOIN users_with_generation u            ON t.USER_ID = u.ID
    GROUP BY u.GENERATION
),

----------------------------------------
-- Sum total sales for Health & Wellness category
----------------------------------------
total_sales_health_wellness AS (
    SELECT
       SUM(HEALTH_WELLNESS_SALES) AS TOTAL_HEALTH_WELLNESS_SALES
    FROM sales_health_wellness
)

SELECT 
   GENERATION
  ,CATEGORY_1
  ,ROUND((a.HEALTH_WELLNESS_SALES / b.TOTAL_HEALTH_WELLNESS_SALES) * 100, 2) AS PERCENTAGE_OF_SALES

FROM sales_health_wellness a
  CROSS JOIN total_sales_health_wellness b

"""
```

```python
# SECOND: Provide SQL Queries Open-ended Questions

# 2d: Who are Fetch’s power users?
### ANSWER: Millennial Females are Fetch's power users by sales $
two_d_query = """

----------------------------------------
-- Create GENERATION based on birth year
----------------------------------------
WITH users_with_generation AS (
    SELECT 
       u.ID
      ,u.GENDER
      ,CAST(strftime('%Y', u.BIRTH_DATE) AS INT) AS BIRTH_YEAR
      ,CASE WHEN CAST(strftime('%Y', u.BIRTH_DATE) AS INT) < 1946 THEN 'Silent Generation'
            WHEN CAST(strftime('%Y', u.BIRTH_DATE) AS INT) BETWEEN 1946 AND 1964 THEN 'Baby Boomers'
            WHEN CAST(strftime('%Y', u.BIRTH_DATE) AS INT) BETWEEN 1965 AND 1980 THEN 'Gen X'
            WHEN CAST(strftime('%Y', u.BIRTH_DATE) AS INT) BETWEEN 1981 AND 1996 THEN 'Millennial'
            WHEN CAST(strftime('%Y', u.BIRTH_DATE) AS INT) BETWEEN 1997 AND 2012 THEN 'Gen Z'
            WHEN CAST(strftime('%Y', u.BIRTH_DATE) AS INT) >= 2013 THEN 'Gen Alpha'
            ELSE NULL END AS GENERATION
    FROM users u
)

SELECT 
   GENDER
  ,GENERATION
  ,SUM(FINAL_SALE) AS SALES
FROM Transactions t
  INNER JOIN users_with_generation u ON t.USER_ID = u.ID
GROUP BY GENDER,GENERATION
ORDER BY SUM(FINAL_SALE) DESC
LIMIT 1

"""

# 2e: Which is the leading brand in the Dips & Salsa category?
### ANSWER: TOSTITOS with $260.99 in Dips & Salsa sales
two_e_query = """

----------------------------------------
-- Distinct list of BARCODES by BRAND to remove dupes
----------------------------------------
WITH distinct_barcode_by_brand AS (
    SELECT DISTINCT
       BARCODE
      ,BRAND
    FROM products
    WHERE CATEGORY_1 = "Snacks"
      AND CATEGORY_2 = "Dips & Salsa"
    
)

SELECT 
   BRAND
  ,SUM(FINAL_SALE) AS SALES
FROM transactions t
  INNER JOIN distinct_barcode_by_brand p ON t.BARCODE = p.BARCODE
GROUP BY BRAND
ORDER BY SUM(FINAL_SALE) DESC
LIMIT 1

"""

# 2f: At what percent has Fetch grown year over year?
### ANSWER: Because there is only transaction data for 2024, I am looking at user account creation information
### User growth saw a significant surge from 2015 to 2020, with especially large increases between 2019 and 2021. 
### Growth slowed in 2022 and has been declining since, with a 42.31% drop in 2023 and a 24.79% decrease in 2024.
two_f_query = """

----------------------------------------
-- Generate ACCOUNT_AGE by comparing Current Date with CREATED_DATE
----------------------------------------
WITH users_with_created_year AS (
    SELECT
       u.ID
      ,u.CREATED_DATE
      ,strftime('%Y', CREATED_DATE) AS CREATED_YEAR
     
    FROM users u
),

----------------------------------------
-- Count users by when they created their account with Fetch
----------------------------------------
user_count_by_created_year AS (

    SELECT DISTINCT
       CREATED_YEAR
      ,COUNT(ID) AS COUNT_OF_USERS
      
    FROM users_with_created_year
    GROUP BY CREATED_YEAR

),

----------------------------------------
-- Return Users by year and return the previous year new users
----------------------------------------
yearly_user_growth AS (
    SELECT 
       CREATED_YEAR
      ,COUNT_OF_USERS
      ,LAG(COUNT_OF_USERS) OVER (ORDER BY CREATED_YEAR) AS PREVIOUS_YEAR_USERS
    FROM user_count_by_created_year
)

SELECT 
   CREATED_YEAR AS YEAR
  ,COUNT_OF_USERS
  ,PREVIOUS_YEAR_USERS
  ,ROUND(((COUNT_OF_USERS + 0.00 - PREVIOUS_YEAR_USERS + 0.00) / PREVIOUS_YEAR_USERS + 0.00) * 100, 2) AS YOY_GROWTH_PERCENT
FROM yearly_user_growth
WHERE PREVIOUS_YEAR_USERS IS NOT NULL
ORDER BY YEAR DESC

"""
```
